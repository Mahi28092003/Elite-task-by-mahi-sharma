```python
df["Total_Cost"] = df["Production_Cost"] + df["Distribution_Cost"] + df["Overheads"]
```

```python
df["Retail_Price"] = df["Total_Cost"] * (1 + markup_percentage)
markup_percentage=0.30
```

```python
print(df["Retail_Price"])
```

```python
df["Profit_Margin_%"] = ((df["Retail_Price"] - df["Total_Cost"]) / df["Retail_Price"]) * 100
print(df["Profit_Margin_%"])
```

```python
df["Price_vs_Competitor"] = df["Retail_Price"] - df["Competitor_Price"]
print(df["Price_vs_Competitor"])
```

```python
plt.figure(figsize=(12,6))
x = range(len(df["Product"]))
plt.bar(x, df["Retail_Price"], width=0.4, label="Our Retail Price", align="center")
plt.bar([i+0.4 for i in x], df["Competitor_Price"], width=0.4, label="Competitor Price", align="center")
plt.xticks([i+0.2 for i in x], df["Product"], rotation=75)
plt.ylabel("Price (₹)")
plt.title("Comparison: Our Retail Price vs Competitor Price")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Profit_Margin_%"], bins=10, edgecolor="black")
plt.xlabel("Profit Margin (%)")
plt.ylabel("Number of Products")
plt.title("Profit Margin Distribution Across Products")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.bar(df["Product"], df["Price_vs_Competitor"], color=["green" if x>0 else "red" for x in df["Price_vs_Competitor"]])
plt.xticks(rotation=75)
plt.axhline(0, color="black", linewidth=1)
plt.ylabel("Price Difference (₹)")
plt.title("Price Difference: Our Price vs Competitor")
plt.tight_layout()
plt.show()
```
